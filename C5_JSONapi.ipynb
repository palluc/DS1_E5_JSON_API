Visit the site www.pokeapi.co  (provides a free API to practice consuming info in json format)  

Extract the first 20 pokémon characters (by default you get 20 pokémons), when calling the pokemon endpoint, as in:  
https://pokeapi.co/api/v2/pokemon  
for each Pokémon character you must also obtain its height and weight. The height and weight information can be gathered by using the detail pokemon url, as in: https://pokeapi.co/api/v2/pokemon /{id}  

As you get the data for each Pokémon, store it in a list, where every element is a tuple of this form: (name, height, weight)  

Then use the list to load a pandas data frame

```python
# Import libraries
import requests
import json
import pandas as pd
```

```python
# Call api
req = requests.get("https://pokeapi.co/api/v2/pokemon")
```

```python
#Extract the first 20 pokemon characters using the api to retrieve the URL, name, and url id number

pokemon_list = []
if req.status_code == 200:
    print('Successful GET request')
    print('Content-Type = ', req.headers['Content-Type']) 
    print('Content-Type = ', req.headers['Date']) 
    
    
    #Unserialize text returned
    data = json.loads(req.text)
    
    #returns a list
    characters = data['results']

    #for each element in the list
    for pokemon in characters:
        num = pokemon['url'].split('/')[6]
        #print('Name:', pokemon['name'])
        #print('URL and num:', pokemon['url'],num)
        tuple = (pokemon['name'], pokemon['url'], num)
        print(tuple)
        pokemon_list.append(tuple)
  
    #print(req.text)
else:
    print('Unsuccessful GET request, code: ', req.status_code)
    
    
    
```

```python
# For each Pokémon character obtain the height and weight in addition to its name.
pokemon_data_list = []
for pokemon in pokemon_list:
    poke_name, poke_url, poke_num = pokemon
    #print('Calling api for pokemon: ', poke_name, poke_url)
    req = requests.get(poke_url)
    if req.status_code == 200:
        poke_data = json.loads(req.text)
        pokemon_name = poke_name
        pokemon_ht   = poke_data['height']
        pokemon_wt   = poke_data['weight']
        #print('here', poke_data['height'], poke_data['weight'])
        pokemon_tuple = (pokemon_name, pokemon_ht, pokemon_wt)
        pokemon_data_list.append(pokemon_tuple)
        print(pokemon_data_list)
    else:
        print('Unsuccessful GET request for pokemone: ', poke_name)
        
        
        
    
    
    
```

```python
# load a pandas data frame
df_pokemon = pd.DataFrame(pokemon_data_list, columns=['Name', 'Weight', 'Height'])
df_pokemon
```
